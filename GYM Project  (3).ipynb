Histogram

```python
import matplotlib.pyplot as plt

# Example: Age distribution
df['Age'].hist()
plt.hist(df['Age'],color='#9BA2FF',edgecolor='#0F0F0F')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Example: Gender distribution

df['gender'].value_counts().plot(kind='bar')
plt.hist(df['Age'],color='orange',edgecolor='#227B94')
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

Box Plot

```python
import seaborn as sns

# Example: Age distribution
sns.boxplot(x=df['Age'])
plt.title('Age Box Plot')
plt.xlabel('Age')
plt.show()
```

```python
# Example: Relationship between age and visit_per_week
plt.scatter(df['Age'], df['visit_per_week'])
plt.title('Age vs Visit per Week')
plt.xlabel('Age')
plt.ylabel('Visit per Week')
plt.show()
```

KDE - Kernel Density Estimate

```python
# Example: KDE plot for Age
import warnings
warnings.filterwarnings('ignore')
sns.kdeplot(df['Age'], shade=True)
plt.title('Age KDE Plot')
plt.xlabel('Age')
plt.ylabel('Density')
plt.show()
```

```python
hm=40
```

```python
t_stat,p_value=ttest_1samp(df['Age'],hm)
alpha=0.05

print(f"t-statistic:{t_stat}")
print(f"p-value:{p_value}")

if p_value < alpha:
    print(f"Reject the null hypothesis at alpha = {alpha}")
else:
    print(f"fail to reject the null hypothesis at alpha={alpha}")
```
